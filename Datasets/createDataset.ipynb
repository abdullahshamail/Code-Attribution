```python
import math
ext = '.cpp'
count = 1
for index in range(df.shape[0]):
    if not os.path.isdir(path+str(df['username'][index])):
        os.mkdir(path+str(df['username'][index]))
    userpath = path+str(df['username'][index])+'/'
    if ext in df['file'][index]:
        newfilename = str(df['file'][index].replace('.', str(df['solution'][index])+'.'))
        if not os.path.exists(userpath+newfilename):
            f = open(userpath+newfilename, "w")
            if not (str(df['flines'][index]) == 'nan'):
                f.write(df['flines'][index])
                print("file ", count, "written" )
                count += 1
            f.close()
#     print(index)
    
import shutil
for f in os.listdir(path):
    if not (f.startswith('.') == 0 and os.listdir(path+f)):
#         print(f)
        shutil.rmtree(path+f)
#     else: 
#         shutil.rmtree(path+f)
```
